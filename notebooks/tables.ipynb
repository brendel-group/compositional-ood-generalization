```python
import pandas as pd
```

```python
df = pd.read_csv("/home/thaddaus/code/cood/notebooks/table_data.csv")
```

```python
df.head()
```

```python
res = df.groupby(["Name"], as_index=False).agg({
    "MSE_ID": ["mean", "std"], "MSE_all": ["mean", "std"],
    "R2ScoreVW_ID": ["mean", "std"], "R2ScoreVW_all": ["mean", "std"],
    # "R2Score_ID": ["mean", "std"], "R2Score_all": ["mean", "std"],
})
format_mapping = {
    ("MSE_ID", "mean"): "{:.2e} &",
    ("MSE_ID", "std"): "\\sd{{{:.2e}}} &",
    ("MSE_all", "mean"): "{:.2e} &",
    ("MSE_all", "std"): "\\sd{{{:.2e}}} &",
    ("R2ScoreVW_ID", "mean"): "{:.3f} &",
    ("R2ScoreVW_ID", "std"): "\\sd{{{:.2e}}} &",
    ("R2ScoreVW_all", "mean"): "{:.3f} &",
    ("R2ScoreVW_all", "std"): "\\sd{{{:.2e}}}",
}
res.style.format(format_mapping)
```
